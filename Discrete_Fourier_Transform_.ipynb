```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
n = 1024
j = (np.sqrt(-1 + 0j))
w = np.exp(-j * 2 * np.pi / n)

J, K = np.meshgrid(np.arange(n), np.arange(n))
DFT = np.power(w, J*K)
DFT = np.real(DFT)

plt.imshow(DFT)
plt.show()
```
